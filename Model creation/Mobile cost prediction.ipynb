## Label enconding the categorical values

```python
from sklearn import preprocessing
```

```python
lel=[]
obj=(df.dtypes=='object')
for col in list(obj[obj].index):
    le=preprocessing.LabelEncoder()
    le.fit(df[col])
    lel.append(le)
    df[col]=le.transform(df[col])
```

```python
df.info()
```

```python
arr=(df['Price_in_India'])
df.drop(arr[arr<3000],inplace=True)
df.shape
```

## Splitting the data frame into x and y

```python
x=df.drop(['Price_in_India'],axis=1)
x.shape
```

```python
y=df['Price_in_India']
y.shape
```

## Using train_test_split splitting the data set into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

# RANDOM FOREST

## Creating a random forest model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100,criterion = "absolute_error",random_state=42)
```

```python
rfmodel=rf.fit(X_train,y_train)
```

```python
y_predrf=rfmodel.predict(X_test)
```

```python
mean_absolute_error(y_test,y_predrf)
```

```python
r2_score(y_test,y_predrf)
```
